```python
alunos_db = [
{"matricula": "86534135","nome": "Vinícius", "Curso":"TI"},
{"matricula": "86534136","nome": "Pedro", "Curso":"ADS"},
{"matricula": "86534137","nome": "Ruan", "Curso":"TA"}
]
print ("1. SISTEMA SIMPLIFICADO DE CADASTRO DE ALUNOS")
print ("-"*40)
# 1 ADICIONAR NOVO ALUNO
print ("-> adicionando novo aluno...")
novo_aluno = {"matricula": "86534138","nome": "Vini", "Curso":"INGLES"} # Removed the trailing comma
alunos_db.append(novo_aluno)
print (f"Aluno {novo_aluno['nome']} adicionado com sucesso")
# 2 Buscar aluno por mátricula
matricula_busca = "86534135"
#Acessa os valores
aluno_encontrado = next((aluno for aluno in alunos_db if aluno ["matricula"]==matricula_busca), None)
if aluno_encontrado:
    print (f"\nAluno encontrado: (matricula: {matricula_busca})")
    #acessa os valores
    print(f"nome: {aluno_encontrado['nome']}, Curso: {aluno_encontrado['Curso']}")
else:
    print(f"Aluno com matricula {matricula_busca} não encontrado.")

# 3 Atualizar informações do curso (UPDATE)
matricula_update = "854485674"
for aluno in alunos_db:
    aluno ["matricula"] == matricula_update
    aluno ["Curso"] = "Engenharia"
    print(f"\nCurso do aluno {aluno['nome']} atualizado para {aluno['Curso']}")
    break
print (f"Aluno: {aluno['nome']}, Curso: {aluno['Curso']}")
print ("-"*40)
```
