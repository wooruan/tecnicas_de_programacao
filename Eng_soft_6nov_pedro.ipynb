```python
# 4.Solicite que o usuário digite seu nome.
# Solicite que o usuários digite seu ano de nascimento (formato AAAA).
# Exiba na tela:
# nome_usuario nasceu em ano_nascimento!
# (nome_usuario e ano_nascimento
# são as variáveis)

nome = str(input("digite seu nome: "))
nasc = str(input("digite sua data de nascimento"))

print("o usuario: "+ nome+ " nasceu em: "+ nasc)
```

```python
# 5.Solicite que o usuário digite seu nome e sua idade.
# Imprima “Olá, nome! Você tem idade anos?”
# Imprima o ano de nascimento da pessoa (ignore meses e dias)
# Imprima o nome da pessoa idade vezes (João tem 18, imprime JoãoJoão...)

nome = str(input("digite seu nome: "))
idade = int(input("digite sua idade: "))
ano = 2025
idade = ano - idade
print (nome, " nasceu em: ", idade, nome*idade)
```

```python
# 6. Solicite ao usuário que digite três notas
# depois calcule e exiba a média ponderada

nota1 = float(input("digite a primeira nota: "))
nota2 = float(input("digite a segunda nota: "))
nota3 = float(input("digite a terceira nota: "))

mp = ((nota1 * 1) + (nota2*2) + (nota3*3))/(1+2+3)
print("media ponderada: ", mp)
```

```python
# 7. Solicite ao usuário três notas e seus
# respectivos pesos, depois calcule e exiba a
# média ponderada

nota1 = float(input("digite a primeira nota: "))
nota2 = float(input("digite a segunda nota: "))
nota3 = float(input("digite a terceira nota: "))

peso1 = float(input("digite o primeira peso: "))
peso2 = float(input("digite o segundo peso: "))
peso3 = float(input("digite o terceiro peso: "))

mp = ((nota1 * peso1) + (nota2*peso2) + (nota3*peso3))/(1+2+3)
print("media ponderada: ", mp)
```

```python
# 8. Peça ao usuário um número e exiba a soma
# de seus dígitos.

num = input("digite um numero: ")
nuum = sum(map(int, num))
print(nuum)
```
