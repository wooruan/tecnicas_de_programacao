```python
'''Escreva um programa que solicite o valor fixo do salário de um vendedor, o total vendido no mes e o percentual de comissao do vendedor. No final apresentar o salario bruto'''
salario = float(input("Digite o valor fixo do salário do vendedor: "))
total_vendido = float(input("Digite o total vendido no mês: "))
percentual_comissao = float(input("Digite o percentual de comissão do vendedor: "))
salario_bruto = salario + (total_vendido * (percentual_comissao / 100))
print(f"O salário bruto do vendedor é: R$ {salario_bruto:.2f}")

```
